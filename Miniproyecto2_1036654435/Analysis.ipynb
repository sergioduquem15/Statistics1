```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def m_numbers(J):
    return np.arange(-J,J+1)
```

```python
def or_dot(n):
    dot = []
    for i in range(0,n):
        dot.append(i+0.5)
    return dot
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("3D Quiver plot")
n = 1
a = np.arange(0,390,30) # cualquier angulo.
bb = np.arange(0,210,30)
eta = [a*np.pi/180 for a in a]
theta = [bb*np.pi/180 for bb in bb]
mag = 0.05

for i in range(n):
    for j in range(n):
        for k in range(n):
            origin = np.array([bb[i],bb[j],bb[k]])
            for rad1 in eta:
                for rad2 in theta:
                #rand1 = np.random.randint(0,360)
                #rand2 = np.random.randint(0,180)
                #eta = rand1*np.pi/180
                #theta = rand2*np.pi/180
                    V = np.array([[mag*np.cos(rad1)*np.sin(rad2) , 
                    mag*np.sin(rad1)*np.sin(rad2),
                    mag*np.cos(rad2)]])
                    #quiver(*origin, V[:,0], V[:,1], V[:,2] ,color=['r','b','g'], scale=10)
                    ax.quiver(*origin, V[:,0], V[:,1], V[:,2] ,color=['r','b','g'])
                    continue
                continue
            continue
#ax.set_xlim(0,n)
#ax.set_ylim(0,n)
#ax.set_zlim(0,n)
plt.grid()
plt.show()
```

```python
from matplotlib import scale


fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
mag = 5
#ax.set_title("3D Quiver plot")
n = 1
m = m_numbers(1/2)
b = or_dot(n)
all_eta = np.arange(0,390,15)
conv_eta = []
for i in all_eta:
    conv_eta.append(i*np.pi/180)
all_theta = np.arange(0,210,15)
conv_theta = []
for i in all_theta:
    conv_theta.append(i*np.pi/180)

for i in range(n):
    for k in range(n):
        for rand1 in conv_eta:
            for rand2 in conv_theta:
                origin = np.array([b[i],b[k]])
                #rand1 = np.random.randint(0,360)
                rand3 = np.random.randint(0,len(m))
                #eta = rand1*np.pi/180
                #theta = rand2*np.pi/180
                V = np.array([[mag*np.cos(rand1)*np.sin(rand2), mag*m[rand3]]])
                #plt.quiver(*origin, V[:,0], V[:,1] ,color=['r','b','g'])
                ax1.quiver(*origin, V[:,0], V[:,1] ,color=['r','b','g'],scale=25, width=0.001, pivot='middle')
                continue
        continue
ax1.set_xlim(0,n)
ax1.set_xlabel("Eje X")
ax1.set_ylabel("Eje Z")
ax1.set_ylim(0,n)
ax1.grid()

ax2 = fig.add_subplot(212)
for i in range(n):
    for j in range(n):
        for rand1 in conv_eta:
            for rand2 in conv_theta:
                origin = np.array([b[i],b[j]])
                #rand1 = np.random.randint(0,360)
                #rand3 = np.random.randint(0,len(m))
                #eta = rand1*np.pi/180
                #theta = rand2*np.pi/180
                V = np.array([[mag*np.cos(rand1)*np.sin(rand2), mag*np.sin(rand1)*np.sin(rand2)]])
                #plt.quiver(*origin, V[:,0], V[:,1] ,color=['r','b','g'])
                ax2.quiver(*origin, V[:,0], V[:,1] ,color=['r','b','g'],scale=25, width=0.001, pivot='middle')
                continue
        continue
ax2.set_xlim(0,n)
ax2.set_ylim(0,n)
ax2.set_xlabel("Eje X")
ax2.set_ylabel("Eje Y")
ax2.grid()

plt.show()
```
