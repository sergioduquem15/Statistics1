```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
abs_mu = 1
L = 1 # lado de la caja
N = L**3 # numero de momentos magneticos dentro de la caja.
```

```python
def or_dot(n):
    dot = []
    for i in range(0,n):
        dot.append(i+0.5)
    return dot
```

```python
def m_numbers(J):
    return np.arange(-J,J+1)
```

```python
mag = np.sqrt(1/2) # magnitud normalizada
#mag = 1 # magnitud
a = np.arange(0,390,30) # cualquier angulo.
eta = [a*np.pi/180 for a in a]
n = 10

def or_dot(n):
    dot = []
    for i in range(0,n):
        dot.append(i+0.5)
    return dot

b = or_dot(n)

for i in range(n):
    for j in range(n):
        origin = np.array([b[i],b[j]])
        for rad in eta:
            V = np.array([[mag*np.cos(rad) , mag*np.sin(rad)]])
            plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=21)
        continue
plt.ylim(0,n)
plt.xlim(0,n)
plt.show()
```

```python
for i in range(5):
    for j in range(5):
        origin = np.array([b[i],b[j]])
        rand = np.random.randint(0,360)
        eta = rand*np.pi/180
        V = np.array([[mag*np.cos(eta) , mag*np.sin(eta)]])
        plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=10)
        continue
plt.ylim(0,5)
plt.xlim(0,5)
plt.grid()
plt.show()
```
